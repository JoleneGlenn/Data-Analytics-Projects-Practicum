***Calculate the monthly charge for each customer***

```python
# Create a dictionary to hold month keys and values
months = {'jan': 1,'feb': 2,'mar': 3,'apr': 4,'may': 5,'jun': 6, 'jul': 7,'aug': 8,'sep': 9,'oct': 10,'nov': 11,'dec': 12}
```

```python
# Create a function to calculate the charge for minutes used each month
def mins_charge(row, month_num):
    month_mins = 'mins_{}'.format(month_num)
    if row['reg_date'].month <= month_num:
        extra_mins_act = (row[month_mins] - row['minutes_included']) * row['usd_per_minute']
        if extra_mins_act >= 0:
            return extra_mins_act
        else:
            return 0
    else:
        return 0
    
for month in months.keys():
    month_value = '{}_mins_charge'.format(month)
    user_usage_data[month_value] = user_usage_data.apply(mins_charge, args=(months[month],), axis=1) # args requires a tuple

print('Table title: Client Usage Data\nContains: user info, minutes/month, texts/month, gbs/month, plan details, minutes charged/month')
user_usage_data.head()
```

```python
# Create a function to calculate the charge for texts sent each month
def texts_charge(row, month_num):
    month_texts = 'texts_{}'.format(month_num)
    if row['reg_date'].month <= month_num:
        extra_texts_act = (row[month_texts] - row['messages_included']) * row['usd_per_message']
        if extra_texts_act >= 0:
            return extra_texts_act
        else:
            return 0
    else:
        return 0

for month in months.keys():
    month_value = '{}_texts_charge'.format(month)
    user_usage_data[month_value] = user_usage_data.apply(texts_charge, args=(months[month],), axis=1) # args requires a tuple

print('Table title: Client Usage Data\nContains: minutes/month, texts/month, gbs/month, plan details, minutes charged/month, texts charged/month')
user_usage_data.head()
```

```python
# Create a function to calculate the charge for gbs used each month
def gbs_charge(row, month_num):
    month_gbs = 'gbs_{}'.format(month_num)
    if row['reg_date'].month <= month_num:
        extra_gbs_act = (row[month_gbs] - row['gb_per_month_included']) * row['usd_per_gb']
        if extra_gbs_act >= 0:
            return extra_gbs_act
        else:
            return 0
    else:
        return 0

for month in months.keys():
    month_value = '{}_gbs_charge'.format(month)
    user_usage_data[month_value] = user_usage_data.apply(gbs_charge, args=(months[month],), axis=1) # args requires a tuple
    
print('Table title: Client Usage Data\nContains: user info, minutes/month, texts/month, gbs/month, plan details, minutes charged/month, texts charged/month, gbs charged/month')
user_usage_data.head()
```

```python
# Create a function that calculates the total monthly charge for each client
# Need to account for active and inactive months
import numpy as np
def monthly_total(row, month_name, month_num):
    mins_monthly_charge = '{}_mins_charge'.format(month_name)
    texts_monthly_charge = '{}_texts_charge'.format(month_name)
    gbs_monthly_charge = '{}_gbs_charge'.format(month_name)
    if row['reg_date'].month <= month_num and 'NaT' in str(row['churn_date']):
        monthly_total_act = row['usd_monthly_pay'] + row[mins_monthly_charge] + row[texts_monthly_charge] + row[gbs_monthly_charge]
        return monthly_total_act
    elif row['reg_date'].month <= month_num and row['churn_date'].month >= month_num:
        total_monthly_charge_inact = row['usd_monthly_pay'] + row[mins_monthly_charge] + row[texts_monthly_charge] + row[gbs_monthly_charge]
        return total_monthly_charge_inact
    else:
        return 0

for month in months.keys():
    total_monthly_charge = '{}_total'.format(month)
    user_usage_data[total_monthly_charge] = user_usage_data.apply(monthly_total, args=(month, months[month]), axis=1)
    
print('Table title: Client Usage Data\nContains: minutes/month, texts/month, gbs/month, plan details, minutes charged/month, texts charged/month, gbs charged/month, monthly total charge')
user_usage_data.head(7)
```

***Info on testing hypothesis***

A test of a statistical hypothesis always begins with a statement of the hypothesis. First, we state the **null hypothesis**, **H₀**. For example, "The mean of the population in question equals *A*," where *A* is some number. The **alternative hypothesis**, **H₁**, is based on H₀. For this H₀, H₁ would be, "The mean of the population does not equal *A*." H₀ is always stated with an equal sign.

We construct a distribution based on the assumption that H₀ is true. In this case, it would be a normal distribution around the parameter of interest, the mean. The variance or its square root, the standard deviation (a measure of the distribution's dispersion, or scatter), is estimated based on the sample.

Hypothesis on the Equality of the Means of Two Populations

Sometimes you need to compare the means of two different statistical populations. In order to test your hypothesis that the means of two statistical populations are equal based on samples taken from them, apply the method `scipy.stats.ttest_ind()`. The method takes the following parameters: 

- `array1`, `array2` are arrays containing the samples.
- `equal_var` is an optional parameter that specifies whether or not the variances of the populations should be considered equal.

If there is reason to believe that the samples were taken from populations with similar parameters, then set `equal_var = True`, and the variance of each sample will be estimated from the *combined* dataset from the two samples, not *separately* from the values in each sample. This gives us more accurate results. However, we only do this if the variances of the statistical populations from which the samples are taken are approximately equal. Otherwise, we need to set `equal_var = False`*;* by default, it's Hypothesis on the Equality of the Means of Paired Samples `equal_var = True`.
