```python
# create a first session month column and change datatypes
first_month_info = visits.groupby('user_id').agg({'session_date_D': 'min'}).reset_index()
first_month_info.columns = ['user_id', 'first_session_date']
first_month_info['first_session_month'] = first_month_info['first_session_date'].astype('datetime64[M]')
first_month_info.head()
```

```python
# create a table to use for calculating retention info
retention_info = pd.merge(visits, first_month_info, on = 'user_id')
retention_info.head()
```

```python
# calculate cohort age
retention_info['cohort_age'] = retention_info['session_month'] - retention_info['first_session_month']
retention_info['cohort_age'] = ((retention_info['cohort_age'] / np.timedelta64(1, 'M')).round()).astype('int')
retention_info.head()
```
